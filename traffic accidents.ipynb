```python
import pandas as pd
import numpy as np
import folium
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("NYC Accidents 2020.csv")
df.head()
```

```python
nan_locations = df.isna().any()
nan_locations
```

```python
nyc_map = folium.Map(location=[40.7128,  -74.0060], zoom_start=12)
nyc_map
```

```python
missing_data_numbers = df.isnull().sum()
missing_data_numbers
```

```python
df.describe()
```

```python
df.info()
```

```python
missing_data = df.isnull().sum() / len(df) * 100
print(missing_data.sort_values(ascending=False))
```

```python
threshold = 73.0
df = df.loc[:, missing_data < threshold]
```

```python
df.info()
```

```python
df['LOCATION'] = df.apply(
    lambda row: f"{row['LATITUDE']}, {row['LONGITUDE']}" if pd.isnull(row['LOCATION']) else row['LOCATION'],
    axis=1
)
```

```python
df.sample(5)
```

```python
df = df.dropna(subset=['LATITUDE', 'LONGITUDE', 'CONTRIBUTING FACTOR VEHICLE 2', 'ZIP CODE', 'ON STREET NAME', 'CROSS STREET NAME', 'CONTRIBUTING FACTOR VEHICLE 2', 'VEHICLE TYPE CODE 2'])
```

```python
df['MISSING_BOROUGH'] = df['BOROUGH'].isnull().astype(int)
```

```python
df.sample(7)
```

```python
df.info()
```

```python

# Loop through the data to add markers
for index, row in df.iterrows():
    # Set marker coordinates
    latitude = row['LATITUDE']
    longitude = row['LONGITUDE']
    
    # Customize popup text with relevant details
    popup_text = (f"<b>Crash Date:</b> {row['CRASH DATE']}<br>"
                  f"<b>Crash Time:</b> {row['CRASH TIME']}<br>"
                  f"<b>Borough:</b> {row['BOROUGH']}<br>"
                  f"<b>Street Name:</b> {row['ON STREET NAME']}<br>"
                  f"<b>Number Injured:</b> {row['NUMBER OF PERSONS INJURED']}<br>"
                  f"<b>Number Killed:</b> {row['NUMBER OF PERSONS KILLED']}<br>"
                  f"<b>Pedestrians Injured:</b> {row['NUMBER OF PEDESTRIANS INJURED']}<br>"
                  f"<b>Contributing Factor:</b> {row['CONTRIBUTING FACTOR VEHICLE 1']}")

    # Add marker to the map
    folium.Marker(
        location=[latitude, longitude],
        popup=folium.Popup(popup_text, max_width=300),
        icon=folium.Icon(color="red" if row['NUMBER OF PERSONS KILLED'] > 0 else "blue")
    ).add_to(nyc_map)

# Save map as an HTML file
nyc_map.save("nyc_traffic_accidents_map.html")
```
